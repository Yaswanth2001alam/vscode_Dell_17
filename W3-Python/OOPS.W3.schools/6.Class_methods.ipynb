```python
class Person:
  def __init__(self, name):
    self.name = name

  def greet(self):
    print("Hello, my name is " + self.name)

p1 = Person("Emil")
p1.greet()
```

```python
class yaswanth:
    def __init__(self,alam):
        self.alam = alam
    def sai(self):
        print("Hello, I am yaswanth alam " + self.alam)

p1 = yaswanth("network engineer")
p1.sai()
```

```python
class Calculator:
  def add(self, a, b):
    return a + b

  def multiply(self, a, b):
    return a * b

calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(4, 7))
```

```python
class alam:
    def add(self, a , b):
        return a + b
    
    def multiply(self, a , b):
        return a * b
    
calc = alam()
print(calc.add(5,10))
print(calc.multiply(2,4))
```

```python
class vpn:
    def add(self, x , y):
        return x + y

    def sub(self, x , y):
        return x - y

exp = vpn()
print(exp.add(5,8))
print(exp.sub(2,6))
```
